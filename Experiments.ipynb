# Zależność od confident

```python
df_path="data/transcations_max50.npy"
min_support=0.5
# min_confident=1

print(df_path)

time_my_appriori = []
time_apriori = []

list_confident = [x / 10.0 for x in range(1, 11, 1)]

for min_confident in list_confident:
    start = time()
    global_freq_item_set, rules = wrapper_appriori(
        df_path=df_path,
        min_support=min_support,
        min_confident=min_confident,
    )
    end = time()
    print(f"Zaimplementowany Apriori: min support {min_support}, min confident {min_confident}")
    print(f"{end - start} s")
    time_my_appriori.append(end - start)
    print(rules)
    
    print()
    
    if df_path.endswith('csv'):
        item_set_list = get_from_file(df_path)
    else:
        with open(df_path, "rb") as f:
            item_set_list = np.load(f, allow_pickle=True)
    
    itemSetList = [tuple(el_list) for el_list in item_set_list]
    
    start = time()
    itemsets, rules = apriori(itemSetList, min_support=min_support, min_confidence=min_confident)
    end = time()
    print(f"Efficiency Apriori: min support {min_support}, min confident {min_confident}")
    print(f"{end - start} s")
    time_apriori.append(end - start)
    print(rules)
    print("*"*20)
```

```python
fig, ax1 = plt.subplots()

ax1.plot( list_confident, time_my_appriori, label="My Apriori")
ax1.set_xlabel("Confident")
ax1.set_ylabel("Duration (s)")
ax1.plot(list_confident, time_apriori,  label="Eficent Apriori")

plt.title("Zależność od confident")
plt.legend()
plt.show()
fig.savefig('img/transcations_max50_confident.png')
```
